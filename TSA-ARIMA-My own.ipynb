# Time series analysis part

```python
from sklearn.preprocessing import MinMaxScaler
Mn=MinMaxScaler()
ConvertedData=Mn.fit_transform(stock_data1[[column]])
ConvertedData
```

```python
ConvertedData.shape
```

```python
orders=[(1,1,1),(1,1,2),(2,3,1),(2,2,2)]
from statsmodels.tsa.arima.model import ARIMA
for i in orders:
    model=ARIMA(ConvertedData,order=i)
    md=model.fit()
    y_pred=md.predict(0,len(ConvertedData)-1)
    from TSA_functions import rmsemape
    rmsemape(ConvertedData,y_pred)
    print ("##################################")
        
        
        
```

```python
len(y_pred)
```

```python
y_pred
```

```python
Actual=pd.DataFrame(ConvertedData,columns=[column])
Actual
```

```python
Predicted=pd.DataFrame(y_pred,columns=[column])
Predicted
```

```python
org_Actual=Mn.inverse_transform(Actual)
org_Actual=pd.DataFrame(org_Actual,columns=[column])
org_Actual
```

```python
org_Predicted=Mn.inverse_transform(Predicted)
org_Predicted=pd.DataFrame(org_Predicted,columns=[column])
org_Predicted
```

```python
from TSA_functions import graph
graph(org_Actual,org_Predicted,"Actual","Predicted","TataCoffee-Close Actual vs Predicted","Days","Prices")
```
